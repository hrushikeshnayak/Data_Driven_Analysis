# **Project Summary -**

## Analysis of Airbnb Listings in New York City

#### Objective:
The primary goal of this project is to analyze Airbnb listings in New York City . This includes cleaning the dataset, performing statistical analysis, and conducting exploratory data analysis (EDA) to understand the distribution of listings, prices, and other significant factors affecting the Airbnb market in NYC.

#### Data Cleaning:
The dataset "Airbnb NYC 2019" consists of multiple columns including `id`, `name`, `host_id`, `host_name`, `neighbourhood_group`, `neighbourhood`, `latitude`, `longitude`, `room_type`, `price`, `minimum_nights`, `number_of_reviews`, `last_review`, `reviews_per_month`, `calculated_host_listings_count`, and `availability_365`. The initial data cleaning steps involved:

1. **Handling Missing Values:** Missing values in `reviews_per_month` were replaced with 0, and missing values in `last_review` were filled with 'No_reviews'.
2. **Removing Duplicates:** Duplicate rows were removed to ensure data integrity.
3. **Dropping Rows with Remaining Missing Values:** Any remaining rows with missing values were dropped.

#### Statistical Analysis:
Statistical analysis was performed to gain insights into the data:

- **Summary Statistics:** Key metrics like mean, standard deviation, minimum, and maximum values for numerical columns were calculated.
- **Distribution Analysis:** The distribution of prices was visualized to understand the spread and skewness.

#### Exploratory Data Analysis (EDA):
EDA was conducted to explore and visualize key trends and patterns in the dataset:

1. **Boxplots of Numerical Columns:** Boxplots for each numerical column were created to identify outliers and understand data distribution.
2. **Listings per Neighbourhood Group:** A count plot was created to show the number of listings in each neighbourhood group.
3. **Pair Plot:** A pair plot for columns like `price`, `minimum_nights`, `number_of_reviews`, and `reviews_per_month` was created to visualize relationships between variables.
4. **Boxplots of Neighbourhood Group vs Price:** A boxplot was created to compare prices across different neighbourhood groups.
5. **Boxplots of Room Type vs Price:** A boxplot was created to compare prices for different room types.
6. **Availability per Neighbourhood Group:** A boxplot was created to visualize availability across different neighbourhood groups.
7. **Geospatial Distribution of Listings:** A scatter plot of listings based on their latitude and longitude was created to show their geographical distribution.
8. **Categorizing Price and Availability:** Prices were categorized into bins ('low', 'moderate', 'high', 'very high'), and availability was categorized into bins ('low', 'moderate', 'high'). Grouped bar plots were created to show the distribution of these categories.
9. **Top 10 Neighbourhoods with High Availability:** A bar plot was created to show the top 10 neighbourhoods with the highest count in the 'high' availability category.

#### Chart - 6

```python
# Categorizing price into bins
df['price_category'] = pd.qcut(df['price'], q=4, labels=['low', 'moderate', 'high', 'very high'])

df['availability_category'] = pd.cut(df['availability_365'], bins=3, labels=['low', 'moderate', 'high'])

grouped_df = df.groupby(['price_category', 'availability_category']).size().unstack()

ax = grouped_df.plot(kind='bar', stacked=True, figsize=(12,8))
plt.xlabel("Price Category")
plt.ylabel("Count of Listings")
plt.title("Price Category vs Count of Listings for each Availability Category")
plt.legend(title="Availability Category", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

##### 1. Why did you pick the specific chart?

This chart categorizes price and availability into bins, providing a comparison of these categories against the total number of listings.

##### 2. What is/are the insight(s) found from the chart?

It appears that listings in the "very high" price category tend to have the highest availability, suggesting they might be less frequently booked.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Properties in the "very high" price category could benefit from targeted marketing that emphasizes their luxury features and amenities to attract guests willing to pay a premium.

#### Chart - 7

```python
# Grouping by 'neighbourhood_group' and 'availability_category'
grouped_df_neighbour_availability = df.groupby(['neighbourhood_group', 'availability_category']).size().unstack()

grouped_df_neighbour_availability.plot(kind='bar', stacked=True, figsize=(14,8))
plt.xlabel("Neighbourhood Group")
plt.ylabel("Count of Listings")
plt.title("Distribution of Availability Categories by Neighbourhood Groups")
plt.legend(title="Availability Category", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()

grouped_df_neighbour = df.groupby(['neighbourhood_group', 'price_category']).size().unstack()
grouped_df_neighbour.plot(kind='bar', stacked=True, figsize=(14,8))
plt.xlabel("Neighbourhood Group")
plt.ylabel("Count of Listings")
plt.title("Distribution of Neighbourhood Groups by Price Category ")
plt.legend(title="Price Category", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
availability_counts = df.groupby(['neighbourhood', 'availability_category']).size()

counts_df = availability_counts.reset_index(name='count').sort_values(by='count', ascending=False)


top_10_neighbourhoods = counts_df[counts_df['availability_category'] == 'high'].head(10)

# Plotting the top 10 neighbourhoods with highest count in the 'high' availability category
plt.figure(figsize=(14,8))
sns.barplot(y='neighbourhood', x='count', data=top_10_neighbourhoods, orient='h', palette='viridis')
plt.title('Top 10 Neighbourhoods with "High" Availability Category')
plt.xlabel('Count')
plt.ylabel('Neighbourhood')
plt.show()
```

##### 1. Why did you pick the specific chart?

This visualization provides a concise overview of price and availability distributions, while also highlighting the top 10 neighborhoods with the lowest booking rates.

##### 2. What is/are the insight(s) found from the chart?

Manhattan exhibits the highest availability for expensive stays, followed by Brooklyn. Additionally, the third plot identifies the top 10 neighborhoods with the lowest booking rates, suggesting these areas could benefit from targeted efforts to increase occupancy.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Prioritizing efforts on the top 10 neighborhoods with the lowest booking rates could lead to significant improvements in overall occupancy.
